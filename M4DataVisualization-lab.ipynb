# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connectez-vous à la base de données SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Chargez les données de la colonne ConvertedComp
df = pd.read_sql_query("SELECT ConvertedComp FROM master;", conn)

# Fermez la connexion à la base de données
conn.close()

# Tracer l'histogramme
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=50, color='blue', edgecolor='black')
plt.title('Histogramme de la Compensation Convertie')
plt.xlabel('Compensation Convertie')
plt.ylabel('Fréquence')
plt.grid(axis='y', alpha=0.75)
plt.show()






```

### Box Plots


Plot a box plot of `Age.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connectez-vous à la base de données SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Chargez les données de la colonne Age
df = pd.read_sql_query("SELECT Age FROM master;", conn)

# Fermez la connexion à la base de données
conn.close()

# Tracer le box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'].dropna(), vert=False)  # Box plot horizontal
plt.title('Box Plot de l\'Âge')
plt.xlabel('Âge')
plt.grid(axis='x', alpha=0.75)
plt.show()


```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connectez-vous à la base de données SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Chargez les données de la colonne Age et WorkWeekHrs
df = pd.read_sql_query("SELECT Age, WorkWeekHrs FROM master;", conn)

# Fermez la connexion à la base de données
conn.close()

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5, color='blue')
plt.title('Scatter Plot de l\'Âge et des Heures de Travail par Semaine')
plt.xlabel('Âge')
plt.ylabel('Heures de Travail par Semaine')
plt.grid(alpha=0.75)
plt.show()


```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connectez-vous à la base de données SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Chargez les données nécessaires
df = pd.read_sql_query("SELECT WorkWeekHrs, CodeRevHrs, Age FROM master;", conn)

# Fermez la connexion à la base de données
conn.close()
df = df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])
# Tracer le bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age'].dropna()*10, alpha=0.5, color='blue', edgecolors='w', linewidth=0.5)
plt.title('Bubble Plot de Heures de Travail par Semaine et Heures de Révision de Code')
plt.xlabel('Heures de Travail par Semaine')
plt.ylabel('Heures de Révision de Code')
plt.grid(alpha=0.75)
plt.show()
```

### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python

import sqlite3
import pandas as pd

# Connectez-vous à la base de données SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Chargez toutes les données de la table master
df = pd.read_sql_query("SELECT * FROM master;", conn)

# Fermez la connexion à la base de données
conn.close()

# Affichez les colonnes et un échantillon de valeurs
for column in df.columns:
    print(f"Colonne : {column}")
    print(df[column].dropna().unique()[:10])  # Affichez les 10 premières valeurs non nulles
    print("----")

```

```python

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connectez-vous à la base de données SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Chargez les données nécessaires (modifiez le nom de la colonne si nécessaire)
df = pd.read_sql_query("SELECT UndergradMajor FROM master;", conn)

# Fermez la connexion à la base de données
conn.close()

# Traitez les données pour obtenir les 5 bases de données les plus souhaitées
# Supposons que la colonne soit séparée par des points-virgules
database_counts = df['UndergradMajor'].dropna().str.split('; ', expand=True).stack().value_counts()
top_databases = database_counts.head(5)

# Créez le graphique circulaire
plt.figure(figsize=(10, 6))
plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 des Bases de Données que les Répondants Souhaitent Apprendre l\'Année Prochaine')
plt.axis('equal')  # Égaliser l'axe pour un cercle
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connectez-vous à la base de données SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Chargez les données nécessaires
df = pd.read_sql_query("SELECT Age, WorkWeekHrs, CodeRevHrs FROM master;", conn)

# Fermez la connexion à la base de données
conn.close()

# Filtrer les données pour le groupe d'âge de 30 à 35 ans
age_filtered = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculer les médianes
median_workweek = age_filtered['WorkWeekHrs'].median()
median_coderev = age_filtered['CodeRevHrs'].median()

# Créer un DataFrame pour le graphique
median_data = pd.DataFrame({
    'Category': ['WorkWeekHrs', 'CodeRevHrs'],
    'Median': [median_workweek, median_coderev]
})

# Créer le graphique empilé
plt.figure(figsize=(8, 6))
plt.bar(median_data['Category'], median_data['Median'], color=['blue', 'orange'])
plt.title('Heures Médianes de Travail par Semaine et de Révision de Code (30-35 ans)')
plt.ylabel('Heures Médianes')
plt.show()


```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connectez-vous à la base de données SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Chargez les données nécessaires
df = pd.read_sql_query("SELECT Age, ConvertedComp FROM master;", conn)

# Fermez la connexion à la base de données
conn.close()

# Filtrer les données pour le groupe d'âge de 45 à 60 ans
age_filtered = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculer la médiane
median_converted_comp = age_filtered['ConvertedComp'].median()

# Créer un graphique
plt.figure(figsize=(8, 6))
plt.bar(['45-60 ans'], [median_converted_comp], color='green')
plt.title('Médiane du Salaire Converti pour les Âges de 45 à 60 Ans')
plt.ylabel('Médiane du Salaire Converti')
plt.ylim(0, median_converted_comp * 1.2)  # Ajuster l'échelle pour la clarté
plt.show()


```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connectez-vous à la base de données SQLite
conn = sqlite3.connect('m4_survey_data.sqlite')

# Chargez les données nécessaires
df = pd.read_sql_query("SELECT MainBranch FROM master;", conn)

# Fermez la connexion à la base de données
conn.close()

# Compter les occurrences de chaque catégorie
main_branch_counts = df['MainBranch'].value_counts()

# Créer un graphique à barres horizontal
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.title('Répartition des Principales Branches')
plt.xlabel('Nombre de Répondants')
plt.ylabel('Principale Branche')
plt.grid(axis='x,y')
plt.show()


```
